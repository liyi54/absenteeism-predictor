```python
# We extract the categorical(nominal) variables from the Reason for Absence column
# Since Reason 0 means the employee has no reason for being absent at work. We drop the dummy for that column.
# This will represent the default value if the reasons are not part of reasons 1 to 28

reason_dummies = pd.get_dummies(df['Reason for Absence'], drop_first=True)
```

```python
reason_dummies
```

```python
# We drop the Reason for Absence column as we already have dummies representing each reason category
df = df.drop(['Reason for Absence'], axis=1)
```

```python
df
```

```python
# 28 reasons are quite a lot for us to run our analysis on. So we group common reasons together to reduce the number
# of reasons we use as inputs. It is advisable to group these variables in order to reduce the number of columns in 
# our dataframe

reason_group_1 = reason_dummies.loc[:,1:14].max(axis=1)
reason_group_2 = reason_dummies.loc[:,15:17].max(axis=1)
reason_group_3 = reason_dummies.loc[:,18:21].max(axis=1)
reason_group_4 = reason_dummies.loc[:,22:28].max(axis=1)
```

```python
reason_group_1
```

```python
# Now we concatenate the reason groups to the original dataframe

df = pd.concat((df, reason_group_1, reason_group_2, reason_group_3, reason_group_4), axis=1)
df
```

```python
# Rename the columns 0 1 2 3
df.columns.values
```

```python
cols = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']

df.columns = cols
```

```python
df.head()
```

```python
# Re-order the columns

cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[cols]
```

```python
df.head()
```

```python
# At this point, we create a checkpoint for our code
df_mod = df.copy()
```

```python
df_mod
```

```python
# We need to split the Date column into Day, Month and Year. Before we do this, we confirm the data type of the values
# in the column to ensure they are in the datetime format to aid easy splitting
```

```python
df_mod.info()
```

```python
df_mod['Date'] = pd.to_datetime(df_mod['Date'], format = '%d/%m/%Y')
```

```python
df_mod.info()
```

```python
df_mod
```

```python
month = []

for i in range(700):
    month.append(df_mod['Date'][i].month)
```

```python
df_mod['Month Value'] = month
```

```python
df_mod
```

```python
# To extract the day of the week, we use the weekday method in pandas
df_mod['Date'][0].weekday()
```

```python
# We then create a function to enable us apply this to the entire column Date

def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_mod['Day of the Week'] = df_mod['Date'].apply(date_to_weekday)
```

```python
df_mod
```

```python
df_mod = df_mod.drop(['Date'],axis=1)
```

```python
df_mod
```

```python
# The Education column contains values without numerical meaning. Therefore they are categorical(ordinal) variables
# We will map these values to 0 and 1 using zero as the lowest level of education and 1 for any level above 1 in the
# dataset

pd.unique(df_mod['Education'])
```

```python
df_mod1 = df_mod.copy()
```
