```python
import pandas as pd
```

```python
raw_data = pd.read_csv('original.csv')
raw_data.head()
```

```python
data_no_id = raw_data.drop(['ID'], axis=1)
data_no_id.head()
```

```python
raw_dummies = data_no_id['Reason for Absence']

for i in range(len(raw_dummies)):
    if raw_dummies[i] in range(15):
        raw_dummies[i] = 1
    elif raw_dummies[i] in range(15,18):
        raw_dummies[i] = 2
    elif raw_dummies[i] in range(18,22):
        raw_dummies[i] = 3
    else: 
        raw_dummies[i] = 4

raw_dummies.head()
```

```python
data_no_id['Reason for Absence'] = raw_dummies
```
